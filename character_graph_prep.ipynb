```python
import json
```

```python
with open('twin_peaks_clean_locations_characters.json') as data_file:    
    data = json.load(data_file)
```

```python
data
```

```python
character_list = []
character_dict = {}
for a in data:
    for b in data[a]:
        if data[a][b].get('characters'):
            for c in data[a][b]['characters']:
                if c not in character_list:
                    character_list.append(c)
                    character_dict[c] = {}

character_list.sort()
print character_list
print ""
print character_dict
    
```

```python
for character in character_list:
    if int(character) != 0:
        for episode in data:
            for scene in data[episode]:
                if data[episode][scene].get('characters'):
                    if character in data[episode][scene]['characters']:
                        for other_character in data[episode][scene]['characters']:
                            if other_character != character:
                                #print other_character, "other_character"
                                #print character, "character"
                                if character_dict[character].get(other_character):
                                    character_dict[character][other_character]+=1
                                else:
                                    character_dict[character][other_character]=1

print character_dict
```
