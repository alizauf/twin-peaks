```python
import json
import csv
```

```python
with open('twin_peaks_clean_locations_characters.json') as data_file:    
    data = json.load(data_file)
```

```python
data
```

```python
character_list = []
character_dict = {}
character_dict_names = {}
for a in data:
    for b in data[a]:
        if data[a][b].get('characters'):
            for c in data[a][b]['characters']:
                if c not in character_list:
                    character_list.append(c)
                    character_dict[c] = {}
                    character_dict_names[c] = [data[a][b]['characters'][c]['standard_character_name'],data[a][b]['characters'][c]['character_group'] ]

character_list.sort()
print character_dict_names
    
```

```python
# #this loops through the character dictionary to make each character a key and all the characters they have scenes with + a count, the value
for character in character_list:
    if int(character) != -1:
        for episode in data:
            for scene in data[episode]:
                if data[episode][scene].get('characters'):
                    if character in data[episode][scene]['characters']:
                        for other_character in data[episode][scene]['characters']:
                            if (other_character != character) and (int(other_character) != -1):
                                if character_dict[character].get(other_character) is not None: 
                                    character_dict[character][other_character]+=1
                                else:
                                    character_dict[character][other_character]=1

print character_dict
```

```python
character_dict['120']
```

```python
#graph_data = {"source": character,"target": other_character,"value": character_dict[character][other_character]}
colocation_list_bidirectional = []
for character in character_dict:
    #print character_dict[character]
    for other_character in character_dict[character]:
        #print character_dict[character][other_character]
        if {"source": character,"target": other_character,"value":character_dict[character][other_character]} not in colocation_list_bidirectional:
            colocation_list_bidirectional.append({"source": character,"target": other_character,"value":character_dict[character][other_character]})

print len(colocation_list_bidirectional)
print colocation_list_bidirectional
```

```python
#this gets it into the right format to make a force directed matrix
colocation_list = []
for character in character_dict:
    #print character_dict[character]
    for other_character in character_dict[character]:
        #print character_dict[character][other_character]
        if {"source": int(other_character),"target": int(character),"value":character_dict[character][other_character]} not in colocation_list:
            colocation_list.append({"source": int(character),"target": int(other_character),"value":character_dict[character][other_character]})

print len(colocation_list)
print colocation_list
```

```python
for a in colocation_list:
    print character_dict_names[str(a['source'])][0], "and", character_dict_names[str(a['target'])][0], "have", a['value'], "scenes together."
```

```python
for a in colocation_list_bidirectional:
    print character_dict_names[str(a['source'])][0], "and", character_dict_names[str(a['target'])][0], "have", a['value'], "scenes together."
```

```python
outfile = open('links.json', 'w')
json.dump(colocation_list, outfile, indent=4)
outfile.close()
```
