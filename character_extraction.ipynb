```python
import json
import csv
```

```python
with open('twin_peaks_clean_locations.json') as data_file:    
    data = json.load(data_file)
```

```python
data
```

```python
character_list = []
for a in data:
    for b in a['scenes']:
        if b.get('characters'):
            for c in b['characters']:
                if c not in character_list:
                    character_list.append(c)
print character_list

                
```

```python
with open('characters.csv', 'w') as csvfile:
    fieldnames = ['raw_character', 'character_id', 'character_name']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for a in character_list:
        writer.writerow({'raw_character':a})
```
