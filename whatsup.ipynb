```python
import json
with open('twin_peaks_graph.json') as data_file:    
    data = json.load(data_file)
```

```python
data
```

```python
all_characters_linked = []
for link in data['links']:
    if int(link['source']) not in all_characters_linked:
        all_characters_linked.append(int(link['source']))
    if int(link['target']) not in all_characters_linked:
        all_characters_linked.append(int(link['target']))

all_characters_linked.sort()
        
print all_characters_linked

for a in range(129):
    if a not in all_characters_linked:
        print a
    
   
```
