A doc that just helped me more easily grab the images from the fan wiki.

```python
from bs4 import BeautifulSoup
import urllib
import re
```

```python
def wiki_image_scrape(url):
    wiki = urllib.urlopen(url).read()
    wiki_soup = BeautifulSoup(wiki, 'html.parser')
    images = []
    for img in wiki_soup.find_all('img'):
        images.append(img['src'])
    images
    good_images=[]
    for a in images:
        if re.search('ortal',a) is not None:
            good_images.append(a)
    better_images = []
    for b in good_images:
        if re.search('(.*.jpg)',b) is not None:
            better_images.append(re.search('(.*.jpg)',b).group(1))
    for a in better_images:
        print a
```

```python
wiki_image_scrape('http://twinpeaks.wikia.com/wiki/Main_Characters')
```
