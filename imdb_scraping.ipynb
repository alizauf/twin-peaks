```python
from bs4 import BeautifulSoup
import urllib
import re
import json
```

list of episodes, season 1:
http://www.imdb.com/title/tt0098936/episodes?season=1
http://www.imdb.com/title/tt0098936/episodes?season=2

```python
imdb1 = urllib.urlopen('http://www.imdb.com/title/tt0098936/episodes?season=1').read()
soup = BeautifulSoup(imdb1, 'html.parser')
imdb2 = urllib.urlopen('http://www.imdb.com/title/tt0098936/episodes?season=2').read()
soup2 = BeautifulSoup(imdb2, 'html.parser')


```

```python
episode_links=[]
for link in soup.find_all('a', itemprop="name"):
    episode_links.append('http://www.imdb.com/title/'+ re.search('(t{2}\d{7})', link.get('href')).group(0)+'/fullcredits')
for link in soup2.find_all('a', itemprop="name"):
    episode_links.append('http://www.imdb.com/title/'+ re.search('(t{2}\d{7})', link.get('href')).group(0)+'/fullcredits')
    
```

```python
episode_links
```

```python
#characters and their episodes
tp_characters={}
#episodes and their characters
tp_episodes={}
for episode in episode_links:
    link = urllib.urlopen(episode).read()
    soup = BeautifulSoup(link, 'html.parser')
    for character in soup.find_all("td", class_="character"):
        #This line removes line breaks and extra spaces, cleaning up the character names
        character = re.sub(' +',' ',character.get_text().replace('\n',""))
        #Gets rid of alternate actor spelled name on some character names
        if re.search("(\(.*\))",character) is not None:
            character = character.replace(re.search("(\(.*\))",character).group(0),"")
        #Gets rid of some extra spaces at the end
        character = re.sub(' *$','', character)
        character = re.sub('^ *','', character)
        episode_number = episode_links.index(episode)
        if ('uncredited' not in character) and ('credit only' not in character):
            if character not in tp_characters:
                tp_characters[character]=[episode_number]
            else:
                tp_characters[character].append(episode_links.index(episode))
            if episode_links.index(episode) not in tp_episodes:
                tp_episodes[episode_number] = [character]
            else:
                tp_episodes[episode_number].append(character)

print tp_characters
print tp_episodes
        
```

```python
for a in tp_characters:
    print a, tp_characters[a]
```

```python
for a in tp_episodes:
    print a, tp_episodes[a]
```
