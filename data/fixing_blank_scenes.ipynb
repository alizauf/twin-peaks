```python
import json
import csv
```

```python
with open('twin_peaks_clean_locations_characters.json') as data_file:    
    data = json.load(data_file)
```

```python
#removes everything that is blank
for episode in range(30):
    data[episode]['scenes'] = filter(lambda a: a['heading'] != {} or a['characters'] != [], data[episode]['scenes'] )
```

```python
#renumbers scenes with missing scenes removed
for episode in range(len(data)):
    count = 1
    for a in data[episode]['scenes']:
        a['scene_number'] = count
        count += 1
        
```

```python
location = {u'standard_location_name': u'BLUE PINE LODGE', u'location': u'LAKESHORE BY THE BLUE PINE LODGE', u'int_ext': u'EXT', u'standard_location_id': 1, u'time': u'DAY'}
for episode in range(len(data)):
    for a in data[episode]['scenes']:
        print 'old heading', a['heading']
        if a['heading'].get('standard_location_id') == None or a['heading'] == {u'int_ext': u'INT'} or a['heading'] == {u'int_ext': u'EXT'} or a['heading'] == {}:
            print "meets condition"
            a['heading'] = location
        else:
            location = a['heading']
        print 'new heading', a['heading']
```

```python
outfile = open('all_episodes_blank_scenes_removed.json', 'w')
json.dump(data, outfile, indent=4)
outfile.close()
```
