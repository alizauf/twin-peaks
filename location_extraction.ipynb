```python
import json
import csv
import re
```

```python
with open('twin_peaks.json') as data_file:    
    data = json.load(data_file)
```

```python
data['1']
```

```python
locations=[]
for a in data:
    for b in data[a]:
        if data[a][b]['heading'].get('location'):
            if data[a][b]['heading']['location'] not in locations:
                locations.append(data[a][b]['heading']['location'])
```

```python
locations
```

```python
len(locations)
```

```python
with open('locations.csv', 'w') as csvfile:
    fieldnames = ['raw_location', 'location_id', 'location_name']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for a in locations:
        writer.writerow({'raw_location':a})
```

```python
raw_location_mapping={}
with open('locations_appended.csv', 'rU') as csvfile:
    locations_appended = csv.reader(csvfile)
    for row in locations_appended:
        raw_location_mapping[row[0]]=[row[1], row[2]]

print raw_location_mapping
```

```python
for a in data:
    for b in data[a]:
        if data[a][b]['heading'].get('location'):
            data[a][b]['heading']['standard_location_id'] = raw_location_mapping[data[a][b]['heading']['location']][0] #.encode('utf8', 'replace')
            data[a][b]['heading']['standard_location_name'] = raw_location_mapping[data[a][b]['heading']['location']][1] #.encode('utf8', 'replace')
```
