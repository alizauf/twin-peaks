```python
import json
import csv
import re
```

```python
with open('twin_peaks.json') as data_file:    
    data = json.load(data_file)
```

```python
data
```

```python
locations=[]
for a in data:
    for b in a['scenes']:
        #print b['heading']
        if b['heading'].get('location'):
            #print b['heading']['location']
            if b['heading']['location'] not in locations:
                locations.append(b['heading']['location'])
print locations
```

```python
locations
```

```python
len(locations)
#note this bizarrely increased between feb 10 and mar 1
```

```python
with open('locations.csv', 'w') as csvfile:
    fieldnames = ['raw_location', 'location_id', 'location_name']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for a in locations:
        writer.writerow({'raw_location':a})
```

```python
raw_location_mapping={}
with open('locations_appended_2.csv', 'rU') as csvfile:
    locations_appended = csv.reader(csvfile)
    for row in locations_appended:
        raw_location_mapping[row[0]]=[row[1], row[2]]

master_location_list = []
for a in raw_location_mapping:
    if raw_location_mapping[a][0] not in 'location_id':
        location = {'standard_location_id': int(raw_location_mapping[a][0]), 'standard_location_name': raw_location_mapping[a][1]}
        if location not in master_location_list:
            master_location_list.append(location)

master_location_list
```

```python
for a in data:
    for b in a['scenes']:
        if b['heading'].get('location'):
            print b['heading']['location']
            b['heading']['standard_location_id'] = int(raw_location_mapping[b['heading']['location']][0])
            b['heading']['standard_location_name'] = raw_location_mapping[b['heading']['location']][1]
```
